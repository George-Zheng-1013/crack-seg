```python
from transformers import pipeline
from PIL import Image
import requests
import matplotlib.pyplot as plt
```

```python
# load pipe
image_classifier = pipeline(
    task="zero-shot-image-classification", model="google/siglip-so400m-patch14-384"
)

# load image
image = Image.open(
    r"D:\HP\OneDrive\Desktop\学校\课程\专业课\大数据综合工程设计\crack-seg\算法\bbox_img.png"
)

plt.imshow(image)
```

```python
# inference
outputs = image_classifier(
    image,
    candidate_labels=[
        "thin concrete crack",
        "linear crack on concrete surface",
        "spalling concrete surface",
        "peeling damaged concrete",
        "water seepage stain",
        "damp wall surface",
    ],
)
outputs = [
    {"score": round(output["score"], 4), "label": output["label"]} for output in outputs
]
for output in outputs:
    print(output)
```
